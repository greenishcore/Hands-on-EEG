```python
import pandas as pd
```

```python
train_loss = pd.read_csv("C:\\Users\\a1882\Desktop\EEG\\normal\pic\_acc.csv", header=None)
valid_loss = pd.read_csv("C:\\Users\\a1882\Desktop\EEG\\normal\pic\\_vloss.csv", header=None)
accuracy = pd.read_csv("C:\\Users\\a1882\Desktop\EEG\\normal\pic\\_acc.csv", header=None)
```

```python
print(train_loss)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
hex_d1 = '#0026ff'
hex_d2 = '#ff009a'

plt.figure()
plt.plot(train_loss.iloc[1:100,1],hex_d1,label='train loss')
plt.ylabel('train loss')
plt.xlabel('epoch')
plt.plot(valid_loss.iloc[1:100,1],hex_d2,label='valid loss')
plt.ylabel('valid loss')
plt.xlabel('epoch')
plt.legend()

plt.savefig('hybrid_loss.png')
plt.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline


plt.figure()
plt.plot(accuracy.iloc[1:100,1], color = '#0026ff', label='accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend()
plt.savefig('hybrid_accuracy.png')
plt.show()
```
