```python
import pandas as pd
```

```python
train_loss = pd.read_csv("C:\\Users\\a1882\\Desktop\\EEG\\model_saved\\train_loss.txt", header=None)
valid_loss = pd.read_csv("C:\\Users\\a1882\\Desktop\\EEG\\model_saved\\valid_loss.txt", header=None)
accuracy = pd.read_csv("C:\\Users\\a1882\\Desktop\\EEG\\model_saved\\accuracy_loss.txt", header=None)
```

```python
print(train_loss)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
hex_d1 = '#552a28'
hex_d2 = '#005943'

plt.figure()
plt.plot(train_loss.iloc[1:100,1],hex_d1,label='train loss')
plt.ylabel('train loss')
plt.xlabel('epoch')
plt.plot(valid_loss.iloc[1:100,1],hex_d2,label='valid loss')
plt.ylabel('valid loss')
plt.xlabel('epoch')
plt.legend()

plt.savefig('tranformer_1000_100e_loss.png')
plt.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline


plt.figure()
plt.plot(accuracy.iloc[1:100,1], color = '#550300', label='accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend()
plt.savefig('transformer_1000_100e_accuracy.png')
plt.show()
```
