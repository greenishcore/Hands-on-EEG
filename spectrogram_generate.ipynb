```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy
from scipy import signal
```

```python
data = pd.read_csv("C:\\Users\\a1882\\Desktop\\EEG\\eegdata_3000\\slice_data\\lefthand_zyy_04_epocflex_2023.03.22t16.48.29+08.00.md.bp.csv_0.csv")
```

```python
ch = data.iloc[0,:]
print(ch)
```

```python
freq = 128
t = np.arange(0, 3000/freq , 1/freq)
signal = ch.to_numpy()

freq,t,Sxx = scipy.signal.spectrogram(signal, freq)

amplitude = np.abs(Sxx)

fig = plt.figure(figsize=(10, 10))

plt.pcolormesh(t, freq, np.log(amplitude), shading='gouraud')
plt.axis('off')
fig.patch.set_visible(False)

fig.savefig('C:/Users/a1882/Desktop/EEG/eegdata_3000/spectrogram/spectrogram.png',bbox_inches='tight',pad_inches=0)
```
