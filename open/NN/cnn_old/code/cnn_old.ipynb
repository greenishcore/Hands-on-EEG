```python
import os
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader

class EEGDataset(Dataset):
    def __init__(self, filelist, root_dir,data_label,transform=None):
        self.filelist = filelist
        self.root_dir = root_dir
        self.data_label = data_label
        self.transform = transform


    def __len__(self):
        return len(self.filelist)

    def __getitem__(self, index):
        #print('index:', index, 'len(self.filelist):', len(self.filelist), 'len(self.data_label):', len(self.data_label))
        data_label = self.data_label[index]
        file_path = os.path.join(self.root_dir, self.filelist[index])
        eeg_data = np.array(pd.read_csv(file_path))
        return eeg_data, data_label


eeg_dataset = EEGDataset(file_list, root_dir='C:/Users/a1882/Desktop/zyy' ,data_label=label)
eeg_dataloader = DataLoader(eeg_dataset, batch_size=1, shuffle=True)
```

```python
eeg_dataset = EEGDataset(file_list, root_dir='C:/Users/a1882/Desktop/zyy' ,data_label=label)
eeg_dataloader = DataLoader(eeg_dataset, batch_size=1, shuffle=True)
eeg_validation_dataset = EEGDataset(file_list, root_dir='C:/Users/a1882/Desktop/zyy' ,data_label=label)
eeg_validation_dataloader = DataLoader(eeg_validation_dataset, batch_size=1, shuffle=True)
loss_fn = nn.CrossEntropyLoss()

optimizer = torch.optim.SGD(model.parameters(), lr = 0.001, momentum=0.9)
```

```python
def train_one_epoch(epoch_index, tb_writer):
    running_loss = 0.
    last_loss = 0.

    for i, data in enumerate(eeg_dataloader):
        inputs, labels = data

        optimizer.zero_grad()

        outputs = model(inputs.float())

        loss = loss_fn(outputs, labels)
        loss.backward()

        optimizer.step()

        running_loss += loss.item()
        if i % 1000 == 999:
            last_loss = running_loss / 1000
            print('  batch {} loss: {}'.format(i + 1, last_loss))
            tb_x = epoch_index * len(eeg_dataloader) + i + 1
            tb_writer.add_scalar('Loss/train', last_loss, tb_x)
            running_loss = 0.

    last_loss = running_loss / len(eeg_dataloader) # calculate the last loss of the epoch
    tb_x = epoch_index * len(eeg_dataloader) + i + 1
    tb_writer.add_scalar('Loss/train', last_loss, tb_x)
    print('  epoch {} loss: {}'.format(epoch_index, last_loss))

    return last_loss
```

```python
timestamp = datetime.now().strftime('%Y%m%d_%H%M%S')
writer = SummaryWriter('runs/eeg_yrainer_{}'.format(timestamp))
epoch_number = 0

EPOCHS = 5

best_vloss = 1_000_000.

for epoch in range(EPOCHS):
    print('Epoch {}:'.format(epoch + 1))

    model.train(True)
    avg_loss = train_one_epoch(epoch_number, writer)

    model.train(False)

    running_vloss = 0.0
    for i, vdata in enumerate(eeg_validation_dataloader):
        vinputs, vlabels = vdata
        #vlabels = torch.tensor(vlabels, dtype=float)
        voutputs = model(vinputs.float())
        #voutputs = torch.tensor(voutputs, dtype=float)
        vloss = loss_fn(voutputs, vlabels)
        running_vloss += vloss

    avg_vloss = running_vloss / (i+1)
    print('LOSS train {} valid {}'.format(avg_loss, avg_vloss))
    epoch_number + 1
    writer.add_scalar('Training Loss', avg_loss, epoch_number + 1)
    writer.add_scalar('Validation Loss', avg_vloss, epoch_number + 1)

    writer.flush()

    if avg_vloss < best_vloss:
            best_vloss = avg_vloss
            model_path = 'model_{}_{}'.format(timestamp, epoch_number)
            torch.save(model.state_dict(), model_path)

    epoch_number += 1
```
