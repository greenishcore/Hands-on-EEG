```python
import numpy as np
import pandas as pd
def sliding_window(signal, window_size, step_size):
    n_channels, n_samples = signal.shape
    n_windows = int((n_samples - window_size) / step_size) + 1
    windows = np.zeros((n_channels, window_size, n_windows))
    for i in range(n_windows):
        windows[:, :,i ] = signal.iloc[:, i*step_size:i*step_size+window_size]
    return windows

signal = pd.read_csv("C:\\Users\\a1882\Desktop\EEG\eegdata\\raw\lefthand_zyy_05_epocflex_2023.03.22t16.50.54+08.00.md.bp.csv")
print("åŽŸå§‹ä¿¡å·å½¢çŠ¶ï¼š", signal.shape)
signal = pd.DataFrame(signal)
# å®šä¹‰æ»‘çª—å¤§å°å’Œæ»‘åŠ¨æ­¥é•¿
window_size = 1000
step_size = 100

# å¯¹ä¿¡å·è¿›è¡Œæ»‘çª—å¤„ç†
windows = sliding_window(signal, window_size, step_size)

# è¾“å‡ºæ»‘çª—åŽçš„ä¿¡å·å½¢çŠ¶
print("æ»‘çª—åŽä¿¡å·å½¢çŠ¶ï¼š", windows.shape)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.animation as animation
%matplotlib notebook

fig,ax = plt.subplots()

ims = []
for i in range(windows.shape[2]):
    im = ax.imshow(windows[:,:,i], cmap='Greens', aspect='auto',animated=True)
    if i == 0:
        ax.imshow(windows[:,:,1], cmap='Greens', aspect='auto')
    ims.append([im])
ani = animation.ArtistAnimation(fig, ims, interval=30, blit=True,repeat=False)
fig.colorbar(im, ax=ax)
plt.show()
```
