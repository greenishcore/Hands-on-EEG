```python
import time

def send_data(data):
    num_cols = len(data)
    num_rows = len(data[0])

    for i in range(num_rows):
        row_data = []
        for j in range(num_cols):
            row_data.append(data[j][i])
        yield row_data
        time.sleep(1/128)

def main():
    with open("C:\\Users\\a1882\\Desktop\\EEG\\new_implement\\data\\orgin_raw_data\\lefthand_zyy_04_epocflex_2023.03.22t16.48.29+08.00.md.bp.csv", 'r') as f:
        data = f.readlines()

    # 将数据解析成二维列表
    data = [row.strip().split(',') for row in data]

    # 转置二维列表，变成按列存储的形式
    data = list(map(list, zip(*data)))

    for d in send_data(data):
        print(d)
        # return d


if __name__ == '__main__':
    main()
```

```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import pandas as pd
%matplotlib notebook
%matplotlib inline

# 读取CSV文件
df = pd.read_csv("C:\\Users\\a1882\\Desktop\\EEG\\new_implement\\data\\orgin_raw_data\\lefthand_zyy_04_epocflex_2023.03.22t16.48.29+08.00.md.bp.csv")

# 创建一个Figure对象和一个Axes对象
fig, ax = plt.subplots()

# 定义一个函数，用于在动画中更新数据
def update(i):
    # 清除Axes对象
    ax.clear()
    # 绘制新的数据
    ax.plot(df.iloc[0,0+i:1000+i], df.iloc[1,0+i:1000+i])
    # 设置标题和坐标轴标签
    ax.set_title('Waveform')
    ax.set_xlabel('Time')
    ax.set_ylabel('Value')

# 创建动画对象
ani = animation.FuncAnimation(fig, update, frames=len(df), interval=50)

# 显示动画
plt.show()
```

```python
import itertools

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
%matplotlib notebook

def data_gen():
    for cnt in itertools.count():
        t = cnt / 10
        yield t, np.sin(2*np.pi*t) * np.exp(-t/10.)


def init():
    ax.set_ylim(-1.1, 1.1)
    ax.set_xlim(0, 1)
    del xdata[:]
    del ydata[:]
    line.set_data(xdata, ydata)
    return line,

fig, ax = plt.subplots()
line, = ax.plot([], [], lw=2)
ax.grid()
xdata, ydata = [], []


def run(data):
    # update the data
    t, y = data
    xdata.append(t)
    ydata.append(y)
    xmin, xmax = ax.get_xlim()

    if t >= xmax:
        ax.set_xlim(xmin, 2*xmax)
        ax.figure.canvas.draw()
    line.set_data(xdata, ydata)

    return line,

# Only save last 100 frames, but run forever
anim = animation.FuncAnimation(fig, run, data_gen, interval=100, init_func=init,
                              save_count=100)
plt.show()
```
